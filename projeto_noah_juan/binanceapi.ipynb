```python
from binance.client import Client
import pandas as pd
import MetaTrader5 as mt5
import time
from datetime import datetime
```

```python
client = Client()
```

```python
def getdatamt5(symbol):
    dfdatamt5 = pd.DataFrame(mt5.copy_rates_from(symbol, mt5.TIMEFRAME_M15,time.time(), 38000))
    return dfdatamt5
```

```python
def getdata(symbol, start, period):
    dfdata = pd.DataFrame(client.get_historical_klines(symbol, period, start))
    dfdata = dfdata.iloc[:,:6]
    dfdata.columns = ['Time','Open','High','Low','Close','Volume']
    dfdata.set_index('Time', inplace=False)
    dfdata.index = pd.to_datetime(dfdata.index, unit='ms')
    dfdata['Time'] = pd.to_datetime(dfdata['Time'], unit='ms')
    dfdata.iloc[:,1:] = dfdata.iloc[:,1:].astype(float)
    return dfdata
```

```python
dfbtc15 = getdata('BTCUSDT', '2019-01-01', '15m' )
```

```python
dfbtc1d = getdata('BTCUSDT', '2019-01-01','1d')
```

```python
dfbtc15.to_parquet('dfbtc_15m.parquet')
```

```python
dfbtc1d.to_parquet('dfbtc_1d.parquet')
```
